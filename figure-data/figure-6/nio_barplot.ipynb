```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set()
df = pd.DataFrame()
df.loc["UPBE","val"] = 1.94
df.loc["UPBE0","val"] = 2.45
df.loc["EOM-CCSD","val"] = 2.62
df.loc["LAS-tPBE0","val"] = 4.40
df.loc["UHF","val"] = 7.93
df = df.reset_index().rename({"index":"Method","val":"Band Gap (eV)"},axis="columns")
df["kpoints"] = "1x1x1"
df2 = df.copy()

df = pd.DataFrame()
df.loc["UPBE","val"] = 1.78
df.loc["UPBE0","val"] = 4.65
df.loc["EOM-CCSD","val"] = 4.83
df.loc["LAS-tPBE0","val"] = 5.70
df.loc["UHF","val"] = 8.19
df["kpoints"] = "2x2x2"
df = df.reset_index().rename({"index":"Method","val":"Band Gap (eV)"},axis="columns")
df = pd.concat([df,df2])

palette = sns.color_palette('tab10', len(df))
g = sns.catplot(x="Band Gap (eV)", y="Method", data=df, palette=palette, hue="kpoints", kind="bar",
            height=4.5,aspect=0.65,hue_order=["1x1x1","2x2x2"])
plt.title("NiO Band Gap",fontsize=13)
l1 = plt.axvline(4.3,color="brown",linestyle="--",label="Exp")

catplot_handles = g._legend.legendHandles
g._legend.set(visible=False)
legend2 = plt.legend(title='', loc='upper right', handles=catplot_handles+[l1], labels=["1x1x1","2x2x2","Exp"])

plt.ylabel("")
plt.tight_layout()
# plt.savefig("figures/nio_gap.png",dpi=500)
```

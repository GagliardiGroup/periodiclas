```python
import sys
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
sys.path.append("../../../")
from periodiclas.tools.util import LASdata
sns.set()
```

```python
# DMRG Reference values
DMRGIP = -6.39
DMRGEA = 2.54
```

```python
dpi = 200

plt.rcParams.update({
    'xtick.labelsize': 24,
    'ytick.labelsize': 24,
    'axes.labelsize': 28,
    'figure.dpi': dpi,
    'axes.grid': True,
    'grid.linestyle': '--',
    'figure.figsize':(16,10),
    'axes.titlesize':24
})
```

# Si-Fig 8

```python
plt.figure()

pklfile = './hchain.2.16.pkl'
lasdata = LASdata(pkl_fn=pklfile, pct_pdft=0)
homo_e, homo_k = lasdata.get_homo()
lumo_e, lumo_k = lasdata.get_lumo()
plt.scatter(homo_k, homo_e, label="LASSI VB (2H, 16)", color='#9B59B6', marker='^', s=150)
plt.scatter(lumo_k, lumo_e, label="LASSI CB (2H, 16)", color='#2ECC71', marker='^', s=150)

pklfile = './hchain.2.32.pkl'
lasdata = LASdata(pkl_fn=pklfile, pct_pdft=0)
homo_e, homo_k = lasdata.get_homo()
lumo_e, lumo_k = lasdata.get_lumo()
plt.scatter(homo_k, homo_e, label="LASSI VB (2H, 32)", color='#0072B2', marker='o',  s=150)
plt.scatter(lumo_k, lumo_e, label="LASSI CB (2H, 32)", color='#FF6347', marker='o',  s=150)

plt.axhline(y=DMRGIP, color='blue', linestyle='--', linewidth=2, label='DMRG IP(32H)')  # Blue dashed line
plt.axhline(y=DMRGEA, color='red', linestyle='--', linewidth=2, label='DMRG EA(32H)')   # Red dashed line

plt.ylim(-18, 18)
plt.xlabel(r"$\frac{kd}{2\pi}$")
plt.ylabel("Energy (eV)")
plt.xticks()
plt.yticks()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.16), ncol=3,frameon=True, edgecolor='black', fontsize=22)
plt.tight_layout()
plt.savefig('HChainBandGapConvergencefor2HUC.png', dpi=dpi)
```

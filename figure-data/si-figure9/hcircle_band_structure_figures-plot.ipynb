## LASSI vs. DFT vs. DMRG

```python
import sys
sys.path.append("../../") 
from tools import bandh
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tools.util import load_pkl, LASdata, PeriodicData, DMRGdata
sns.set()
```

```python
numh = 2
las = LASdata(pkl_fn=f"./h32_frag{numh}H_dist14_lsi.pkl",pct_pdft=0)
e,k = las.get_homo()
plt.scatter(k,e,label=f"LASSI VB ({numh}H)",marker="o")
e,k = las.get_lumo()
plt.scatter(k,e,label=f"LASSI CB ({numh}H)",marker="o")

numh = 4
las = LASdata(pkl_fn=f"./h32_frag{numh}H_dist14_lsi.pkl",pct_pdft=0)
e,k = las.get_homo()
plt.scatter(k,e,label=f"LASSI VB ({numh}H)",marker="^")
e,k = las.get_lumo()
plt.scatter(k,e,label=f"LASSI CB ({numh}H)",marker="^")

dmrg = DMRGdata("./hcircle_dmrg32_d14_m2000.csv",pct_pdft=0)
plt.axhline(dmrg.homo,linestyle="--",color="blue",label="DMRG IP")
plt.axhline(dmrg.lumo,linestyle="--",color="red",label="DMRG EA")

plt.ylabel("Energy (eV)")
plt.xlabel("kd/2$\pi$")
plt.legend(bbox_to_anchor=(0.975,1.2),ncols=3)
plt.ylim(-16.5,18)
plt.tight_layout()
plt.annotate("C",(-0.03,-13),c="black",size=24)
# plt.savefig(f"figures/h32_las_band_structures.png",dpi=500)
```

```python
from tools.util import LASdata, DMRGdata, PeriodicData
fnal = "HSE06"
per = PeriodicData(f"./periodic_{fnal}_frag2H_14_k16.csv")
e,k = per.get_homo()
plt.scatter(k,e,label=f"{fnal} VB (2H)",marker="o")
e,k = per.get_lumo()
plt.scatter(k,e,label=f"{fnal} CB (2H)",marker="o")

per = PeriodicData(f"./periodic_{fnal}_frag4H_14_k16.csv")
e,k = per.get_homo()
plt.scatter(k,e,label=f"{fnal} VB (4H)",marker="^")
e,k = per.get_lumo()
plt.scatter(k,e,label=f"{fnal} CB (4H)",marker="^")

dmrg = DMRGdata("./hcircle_dmrg32_d14_m2000.csv",pct_pdft=0)
plt.axhline(dmrg.homo,linestyle="--",color="blue",label="DMRG IP")
plt.axhline(dmrg.lumo,linestyle="--",color="red",label="DMRG EA")

plt.legend(bbox_to_anchor=(0.98,1.2),ncols=3)
plt.ylim(-16,18)

plt.ylabel("Energy (eV)")
plt.xlabel("kd/2$\pi$")
plt.tight_layout()
plt.annotate("D",(-0.03,-15),c="black",size=24)
# plt.savefig(f"figures/h32_{fnal}_band_structures.png",dpi=500)
```
